1. Kada i zašto koristimo regularizaciju? Istražiti različite tipove regularizacije i objasniti ih
tekstualno. Kreativnost je dobrodošla, plus je da u markdownu napišete matematičke
formulacije ili ispišete kod kroz koji objašnjavate.

Regularizacija je dobila ime po tome što joj je cilj da funkciju izlaza učini regularnijom tj. pravilnijom, čime se sprečava preterana prilagođenost modela.
Vuk Batatović - Pronalaženje skrivenog znanja - Elektrotehnički fakultet

Regularizacija služi za menjanje ( povećanje i smanjenje ) težina kako bi klasifikacioni uzorak bio precizniji. 
Odnosno koristi se kako bi sprečila preteranu prilagođenost modela ( eng. overfitting ) dodavajući penale funkciji gubitka ( eng. Loss Function ) pre optimizacije. Penali služe kao "kazna" na različite parametre modela čime se smanjuje prilagođenost modela čineći ga generalnijim. Promena stope regulisanja dovodi do toga da što je veća stopa to je težinski faktor više rasponski limitiran. Postoje tri osnovne tehnike regularizacije, a to su: ***L1*** ili ***Lasso*** koja vrši selekciju i dodeljuje velike težinske vrednosti izabranim ulazima, dok će težine neizabranih biti vrlo male ili će postati nule.   , ***L2*** ili ***Ridge*** kontroliše ulazne vrednosti koje odgovaraju nivou korelacije i ***Elastic Net*** regularizacija koja čini spoj prve dve. Koristi penale od sume apsolutnih i kvadratnih vrednosti zajedno. Najveća razlika između L1 i L2 regularizacije je u načinu dodavanja penala. 

Tradicionalne metode kao što su unakrsna validacija ( eng. Cross-Validation ) ili postepena regresija dobro funkcionišu sa manjim setom podataka, dok Lasso i Ridge tehnike se odlično ponašaju i sa velikom količinom podataka.

2. Objasniti **L1 (Lasso) i L2 (Ridge)** regularizaciju. Kada koristimo jednu, a kada drugu i
po čemu se razlikuju?

**L1 (Lasso)**

L1 regularizacija dodaje koeficijent apsolutne vrednosti magnitude kao penal za funkciju gubitka. Lambda predstavlja regulacioni parametar koji ide kao jedan od inputa modela. Povećavajući njenu vrednost može se redukovati preterana prilagođenost modela i time povećati efekat regularizacije i obrnuto.

```python
from IPython.display import display, Math
display(Math(r'+λ\sum_{j = 1}^p{B_j}'))
```

**L2 (Ridge)**

Ova tehnika regularizacije primorava parametre da budu relativno mali tako što dodaje penal jadnak sumi kvadratne vrednosti koeficijenata funkciji gubitka. Što je veća penalizacija, to su manji koeficijenti. Važno je izabrati pravu vrednost Lambde, jer veoma velika vrednost dodaje previše veliki težinski faktor što dovodi do preterane prilagođenosti modela. 

```python
display(Math(r'+λ\sum_{j = 1}^p{|B_j|}'))
```

5. **Dodatno za najjače**: Proširiti klasu FONISLogisticRegression implementiranu u
materijalima sa radionice tako da podržava različite tipove regularizacije kao što to
radi klasa u sklearnu.

```python
class FONISLogisticRegression:
    # KONSTRUKTOR
    def __init__(self, lr = 0.01, num_item = 100, fit_intercept = True, verbose = False):
        self.lr = lr
        self.num_item = num_item
        self.fit_intercept = fit_intercept
        self.verbose = verbose
    
    # PREDVIDJANJA
    def __sigmoid(self, z):
        return 1/(1 + np.exp(-z))
    
    # FUNKCIJA CILJA
    def __logistic_loss(self, y, h):
        return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
    
    # UCENJE KOEFICIJENATA
    def fit(self, X, y):
        if self.fit_intercept:
            intercept = np.ones((X.shape[0], 1))
            X = np.concatenate((intercept, X), axis = 1)
        
        self.ws = np.zeros(X.shape[1])
        
        for i in range(self.num_item):
            z = np.dot(X, self.ws)
            h = self.__sigmoid(z)
            if(self.verbose == True and i % 100 == 0):
                print(f'Logistic Loss: {self.__loss(y, h)}')
            gradient = np.dot(X.T, (h - y)) / y.size
            self.ws -= self.lr * gradient
    
    # PREDVIDJANJE VEROVATNOCE
    def predict_proba(self, X):
        if self.fit_intercept:
            intercept = np.ones((X.shape[0], 1))
            X = np.concatenate((intercept, X), axis = 1)
        return self.__sigmoid(np.dot(X, self.ws))
    
    # PREDVIDJANJE KLASE
    def predict(self, X, threshold = 0.5):
        return self.predict_proba(X) >= threshold
    
    #REGULARIZACIJE
    #Lasso
    #Ridge
```
