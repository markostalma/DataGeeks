####  4. Dat je proizvoljni np niz dužine n. Kreirati novi niz iste dužine koji sadrži iste brojeve pomnožene prvim n brojevima iz Fibonačijevog niza:

```python
npArr = np.array([5,10,15,20,25,30,35,40,45,50])
a, b = 0, 1
counter = 0
fibArr = np.array([])

while counter < len(npArr):
    fibArr = np.append(fibArr,a)
    a, b = b, a + b
    counter+=1
    
newArr = npArr * fibArr
print('Numpy Array:', npArr, '\nFibonacci Array:', fibArr, '\nNew Numpy Array:',newArr)
```

#### 6. Kreirati dve prazne liste forward i backward . Napisati petlju koja od korisnika traži da unosi elemente u listu (stringove) sve dok ne unese bar tri elementa ili prazan string. Svaki string koji unese korisnik dodati u obe liste tako da forward sadrži stringove redosledom kojim ih je korisnik unosio, a backward obrnuto, to jest od poslednjeg do prvog unetog. Van petlje okrenuti backward listu i i proveriti da li su liste jednake.

```python
forward = []
backward = []
counter = 0
while counter < 3:
    add_in_lst = input("Dodajte Vaš element u listu: ")
    forward.append(add_in_lst)
    backward.append(add_in_lst)
    counter += 1
    if counter == 3:
        print(forward)
        print(backward[::-1])
        
print(forward)
print(backward)
```

#### 7. Kreirati string, proizvoljnu rečenicu . Na osnovu tog stringa, napraviti listu stringova, gde je svaki element jedna reč prvobitne rečenice. Iteriranjem kroz listu, saznati koliko slova a ima u svakoj reči. To zapisati u listu listi sledećeg oblika:

#### 8. Kreirati listu naziva kompanija . Proveriti koji je najduži naziv iz liste i taj naziv zameniti za 'too much':

```python
com_list = ['Google', 'Microsoft', 'Apple', 'Oracle', 'Dell', 'Ibm', 'Nvidia', 'Intel', 'Amd' ]
longest = max(com_list, key=len)
print(longest)

for long, item in enumerate(com_list):
    if longest in item:
        item = 'too much'
        com_list[long] = item
com_list
```

    8.1. Proveriti koji naziv ima najviše slova m i zameniti sva slova 'm' sa 'mam'

```python
com_list = ['Google', 'Microsoft', 'Apple', 'Oracle', 'Dell', 'Ibm', 'Nvidia', 'Intel', 'Amd' ]
counter = 0
m = 'm'
for item in com_list:
    if m in item:
        counter += 1
        print(item)
        char_changes = item.replace('m', 'mam')

com_list, counter, char_changes
```

    8.2. Proveriti koji je najkraći naziv iz liste i zameniti ga tako da bude sa svim velikim slovima:

```python
com_list = ['Google', 'Microsoft', 'Apple', 'Oracle', 'Dell', 'Ibm', 'Nvidia', 'Intel', 'Amd' ]
shortest = min(com_list, key=len)
print(shortest)

for upper, item in enumerate(com_list):
    if shortest in item:
        item = shortest.upper()
        com_list[upper] = item
com_list
```
